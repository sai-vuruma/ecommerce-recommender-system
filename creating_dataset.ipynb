```python
import pandas as pd
clean_df = pd.read_csv('/data/cleaned_data.csv')
clean_df.head()
```

```python
#Creating atomic files
inter_df = clean_df[["user_id", "product_id", "event_time", "event_type"]].copy()
inter_df.rename(columns={
    "user_id": "user_id:token",
    "product_id": "product_id:token",
    "event_time": "event_time:float",
    "event_type": "event_type:float"
    }, inplace=True)

user_df = clean_df[["user_id", "user_session"]].copy()
user_df.rename(columns={
    "user_id": "user_id:token",
    "user_session": "user_session:token"
    }, inplace=True)

item_df = clean_df[["product_id", "category_id", "category_code", "brand", "price"]].copy()
item_df.rename(columns={
    "product_id": "product_id:token",
    "category_id": "category_id:token",
    "category_code": "category_code:token_seq",
    "brand": "brand:token",
    "price": "price:float"
    }, inplace=True)
```

```python
#Storing atomic files
inter_df.to_csv('/data/ecommerce/ecommerce.inter', sep='\t', index=False)
user_df.to_csv('/data/ecommerce/ecommerce.user', sep='\t', index=False)
item_df.to_csv('/data/ecommerce/ecommerce.item', sep='\t', index=False)
```
