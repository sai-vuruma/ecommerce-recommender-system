```python
#Creating clean df to feed the model
clean_in_df = raw_in_df.copy()
```

```python
#Drop rows with null values
clean_in_df.dropna(inplace=True, ignore_index=True)
```

```python
clean_in_df
```

```python
#Data cleaning
#Convert event_time to datetime object
clean_in_df['event_time'] = pd.to_datetime(clean_in_df['event_time']).astype('int64')/1000000000

#Convert event_type column to number ratings
clean_in_df['event_type'] = clean_in_df['event_type'].map({
    'view': 1,
    'remove_from_cart': 2,
    'cart': 3,
    'purchase': 5})

#Replacing dots in category_code column with spaces
clean_in_df['category_code'] = clean_in_df['category_code'].str.replace('.', ' ')

#Sort rows with user_session and event_time
clean_in_df = clean_in_df.sort_values(by=['user_session','event_time']).reset_index(drop=True)

#Drop columns no longer needed
#clean_in_df.drop(columns=['category_id', 'event_type', 'user_session', 'price', 'category_code', 'brand'], inplace=True)

#Drop duplicates
clean_in_df.drop_duplicates(inplace=True)
```
